```python
import json
import pandas as pd
```

```python
directory = 'Data'

train_df = pd.read_csv(f'{directory}/train.csv')
test_df = pd.read_csv(f'{directory}/test/test.csv')
train_sentiment_df = pd.read_csv(f'{directory}/train_sentiment.csv')
test_sentiment_df = pd.read_csv(f'{directory}/test_sentiment.csv')


print(f'Train samples: {len(train_df)}')
print(f'Test samples: {len(test_df)}')
```

```python
train_df = train_df.merge(train_sentiment_df, on='PetID', how='left')
test_df = test_df.merge(test_sentiment_df, on='PetID', how='left')
```

```python
# sample 2000 entries from train and 500 from test
train_sample = train_df.sample(n=2000, random_state=442018)
test_sample = test_df.sample(n=500, random_state=442018)
```

## Saving mydata

```python
train_sample.drop(columns=['RescuerID','State','VideoAmt','Description','Name'], inplace=True)
test_sample.drop(columns=['RescuerID','State','VideoAmt','Description','Name'], inplace=True)
```

```python
train_sample.to_csv(f'myData/train_sample.csv', index=False)
test_sample.to_csv(f'myData/test_sample.csv', index=False)
```

```python
ids_train = train_sample['PetID'].tolist()
ids_test = test_sample['PetID'].tolist()
print(ids_train[:5])
```

```python
# Iterate over images and copy only those in the sampled ids
import os
import shutil

source_dir = f'Data/train_images'
dest_dir = f'myData/train_images_sampled'
os.makedirs(dest_dir, exist_ok=True)
for images_path in os.listdir(source_dir):
    pet_id = images_path.split('-')[0]
    if pet_id in ids_train:
        shutil.copy(os.path.join(source_dir, images_path), os.path.join(dest_dir, images_path))


source_dir = f'Data/test_images'
dest_dir = f'myData/test_images_sampled'
os.makedirs(dest_dir, exist_ok=True)
for images_path in os.listdir(source_dir):
    pet_id = images_path.split('-')[0]
    if pet_id in ids_test:
        shutil.copy(os.path.join(source_dir, images_path), os.path.join(dest_dir, images_path))
```
